```python
## Q2: Extracting Tesla Revenue Data Using Webscraping
url_tesla = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
```

```python
# Define a dictionary with a User-Agent header to mimic a browser
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
```

```python
# Use requests to get the HTML content with the headers
response = requests.get(url_tesla, headers=headers)
html_data = response.text
```

```python
tables = pd.read_html(html_data)
```

```python
tesla_revenue = tables[0]
```

```python
#Question 5: Plot Tesla Stock Graph
import plotly.graph_objects as go
```

```python
def make_graph(dataframe, title):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=dataframe['Date'], y=dataframe['Close'], mode='lines', name='Close Price'))

    fig.update_layout(
        title=title,
        xaxis_title='Date',
        yaxis_title='Stock Price (USD)',
        template='plotly_white',
        height=500
    )

    fig.show()
```

```python
# Plot the Tesla stock data
make_graph(tesla_data, "Tesla Stock Price Over Time")
```

```python
#Question 6: Plot GameStop Stock Graph
make_graph(gme_data, "GameStop Stock Price Over Time")
```
